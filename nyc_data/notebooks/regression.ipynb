```python
xtrain, xtest, ytrain, ytest=train_test_split(trip_model_2020_data, trip_model_2020_target, test_size=0.15)

mlp = make_pipeline(StandardScaler(), MLPRegressor(random_state=1, solver='sgd', tol=1e-1, learning_rate='invscaling', max_iter=10000, hidden_layer_sizes=(150,10)))
mlp.fit(xtrain, ytrain)
print('R-squared (MLP): ', mlp.score(xtrain, ytrain))
mlp_cv_score = cross_val_score(mlp, x, y, cv=10)
print("MLP CV mean score: ", mlp_cv_score.mean())
mlp_ypred = mlp.predict(xtest)
mlp_mse = mean_squared_error(ytest, mlp_ypred)
print("MLP MSE: ", mlp_mse)
print("MLP RMSE: ", mlp_mse*(1/2.0))

# score: R^2
```
