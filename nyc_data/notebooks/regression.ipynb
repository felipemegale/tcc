## Prediction modules

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import cross_validate, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import max_error, mean_absolute_error, mean_squared_error, r2_score, make_scorer
```

## Prediction, 2020

```python
mlp_max_error = make_scorer(max_error)
mlp_mean_absolute_error = make_scorer(mean_absolute_error)
mlp_mean_squared_error = make_scorer(mean_squared_error)
mlp_root_mean_squared_error = make_scorer(mean_squared_error, squared=False)
mlp_r2_score = make_scorer(r2_score)

xtrain, xtest, ytrain, ytest=train_test_split(trip_model_2020_data, trip_model_2020_target, test_size=0.15)

mlp = make_pipeline(StandardScaler(), MLPRegressor(random_state=1, solver='sgd', learning_rate='invscaling', max_iter=50000, hidden_layer_sizes=(150,10)))
mlp.fit(xtrain, ytrain)

print('R-squared (MLP): ', mlp.score(xtrain, ytrain))

mlp_cv_score = cross_validate(mlp, trip_model_2020_data, trip_model_2020_target, cv=10,
                               scoring={'max_error': mlp_max_error,
                                       'mean_absolute_error': mlp_mean_absolute_error,
                                       'mean_squared_error': mlp_mean_squared_error,
                                       'root_mean_squared_error': mlp_root_mean_squared_error,
                                       'r2_score': mlp_r2_score})
print("Max error mean (CV):", mlp_cv_score['test_max_error'].mean())
print("MAE mean (CV):", mlp_cv_score['test_mean_absolute_error'].mean())
print("MSE mean (CV):", mlp_cv_score['test_mean_squared_error'].mean())
print("RMSE mean (CV):",mlp_cv_score['test_root_mean_squared_error'].mean())
print("R-2 mean (CV):", mlp_cv_score['test_r2_score'].mean())

mlp_ypred = mlp.predict(xtest)

# score: R^2
```
