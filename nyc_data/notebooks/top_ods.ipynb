```python
# aggregate all months samples into one pyspark dataframe
uber_data_2020 = jan_sample_2020.union(feb_sample_2020)\
    .union(mar_sample_2020)\
    .union(apr_sample_2020)\
    .union(may_sample_2020)\
    .union(jun_sample_2020)\
    .union(jul_sample_2020)\
    .union(aug_sample_2020)\
    .union(sep_sample_2020)\
    .union(oct_sample_2020)\
    .union(nov_sample_2020)\
    .union(dec_sample_2020)\
    .drop('dispatching_base_num')
```
