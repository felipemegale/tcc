```python
import pandas as pd

data_folder="csv"
data_source_cols=["pickup_datetime", "dropoff_datetime", "PULocationID", "DOLocationID", "SR_Flag"]
```

```python
# get taxi zones to translate pickup and dropoff location IDs to zone names
taxi_zones = pd.read_csv(data_folder+"/taxi+_zone_lookup.csv", usecols=["LocationID","Borough","Zone"])
taxi_zones.sort_values("LocationID")

query = taxi_zones.query("LocationID == 198")
query
```

```python
# get high volume for hire vehicle licenses to translate codes to names
hv_licenses = pd.read_csv(data_folder+"/hvfhs_licenses.csv")
hv_licenses
```

```python
def pre_process_chunk(chunk):
    chunk = chunk[(chunk["PULocationID"] < 264) & (chunk["DOLocationID"] < 264)]
    return chunk
```

```python
# get high volume for hire vehicle data from feb 2019 and filter out the unknown zones (264 and 265)
chunk_list = []
feb_2019_iter = pd.read_csv(data_folder+"/2019/fhvhv_tripdata_2019-02.csv", usecols=data_source_cols, chunksize=1000000)
for chunk in feb_2019_iter:
    chunk = pre_process_chunk(chunk)
    chunk_list.append(chunk)
df = pd.concat(chunk_list)
df.to_csv(data_folder+"/2019/nounknown_fhvhv_tripdata_2019-02.csv")
```
